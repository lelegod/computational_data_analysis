```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_graphviz, DecisionTreeRegressor
from pathlib import Path
import graphviz
```

## Load dataset

```python
path = Path().cwd()
data_file = path.parent / "Data" / "Actors.csv"

T = pd.read_csv(data_file, delimiter = ",")
# One hot encoding of actor
T = pd.get_dummies(T,prefix=['Actor'])
```

> a) Is the described problem a classication or regression problem? Motivate your answer.

> b) Which input variables are categorical and which are continuous?

Growing the tree to the to have one sample in each leaf means that the results will be the same for dealing with the problem as a regression problem and a classification problem.
One needs to be aware of the fact that CARTS in sklearn internally convert the design matrix to floats, meaning that all categorical features must be one hot encoded.

> c) For a categorical variable with k unique categories, what is the number of possible splits into two groups? Note that empty groups are not allowed, and that groupings are commutative in the sense that e.g. the split $\{1,2,3\},\{4,5\}$ is equal to the split $\{4,5\},\{1,3,2\}$.

 $ 2^{k–1} – 1 $

> e) Build a tree that predicts IMDb ratings.

```python
# Assume a regression problem then create a decisiontreeclassifier
dtree=DecisionTreeRegressor()

# Casting response into categorical, as decision trees expects cat response
dtree.fit(T[['Actor_SJ', 'Actor_NC', 'Actor_AP', 'Budget']], T['IMDb'])
```

```python
# create a graph and visualize it in the notebook
graph_Data = export_graphviz(dtree, out_file=None, 
                         feature_names=['Actor_SJ', 'Actor_NC', 'Actor_AP', 'Budget'],  
                         class_names=['2.8', '8.3', '4.0', '2.9', '7.8', '8.1', '3.0'],  
                         filled=True, rounded=True,  
                         special_characters=True)
graph = graphviz.Source(graph_Data)
graph
```

```python
# Assume a classification problem then create a decisiontreeclassifier
dtree=DecisionTreeClassifier()

# Casting response into categorical, as decision trees expects cat response
dtree.fit(T[['Actor_SJ', 'Actor_NC', 'Actor_AP', 'Budget']], pd.Categorical(T['IMDb']).codes)
```

```python
# create a graph and visualize it in the notebook
graph_Data = export_graphviz(dtree, out_file=None, 
                         feature_names=['Actor_SJ', 'Actor_NC', 'Actor_AP', 'Budget'],  
                         class_names=['2.8', '8.3', '4.0', '2.9', '7.8', '8.1', '3.0'],  
                         filled=True, rounded=True,  
                         special_characters=True)
graph = graphviz.Source(graph_Data)
graph
```
